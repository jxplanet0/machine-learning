```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from IPython.display import Image
import pandas as pd
import pydotplus
import graphviz
```

```python
dataset = pd.read_csv('/Users/farrelmanazilin/Document/kuliah/data/ruspini.csv')

train_data = dataset[['X', 'Y']]
train_label = dataset[['CLASS']]

dtc = DecisionTreeClassifier()
dtc.fit(train_data, train_label)

test_data = [[130, 45]]
class_result = dtc.predict(test_data)
print('Class = ', class_result)

acc = dtc.score(train_data, train_label)
err = round((1 - acc) * 100, 2)
print('\n\nError ratio = ', err, '%')

dot_data = export_graphviz(
    dtc,  
    out_file=None,  
    feature_names=["X", "Y"],
    class_names=[str(label) for label in train_label['CLASS'].unique()],  
    filled=True,
    rounded=True, 
    special_characters=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_png("decision_tree.png")
Image(graph.create_png())
```

```python
dataset_train = pd.read_csv('/Users/farrelmanazilin/Document/kuliah/data/milk_training.csv')

train_data = dataset_train.iloc[:, :-1]
train_label = dataset_train.iloc[:, -1]

dataset_test = pd.read_csv('/Users/farrelmanazilin/Document/kuliah/data/milk_testing.csv')

test_data = dataset_test.iloc[:, :-1]
test_label = dataset_test.iloc[:, -1]

dtc = DecisionTreeClassifier()
dtc.fit(train_data, train_label)

class_result = dtc.predict(test_data)
print('Class = ', class_result)

acc = dtc.score(test_data, test_label)
err = 1 - acc
print('\n\nError ratio = ', err)
```

```python
feature_names = dataset_train.columns[:-1].tolist()
dot_data = export_graphviz(
    dtc,  
    out_file=None,
    feature_names=feature_names,
    class_names=[str(label) for label in train_label.unique()],
    filled=True,
    rounded=True,
    special_characters=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_png("decision_tree.png")
Image(graph.create_png())
```

```python
kNN = KNeighborsClassifier(n_neighbors=3, weights='distance')
kNN.fit(train_data, train_label)  
y_pred = kNN.predict(test_data)
accuracy = accuracy_score(test_label, y_pred)
error_ratio_knn = 1 - accuracy

print("Akurasi ratio untuk metode KNN Decision Tree: {:.10f}".format(accuracy))
print("Error ratio untuk metode KNN Decision Tree: {:.10f}".format(error_ratio_knn))
```

```python
classifier = GaussianNB()
classifier.fit(train_data, train_label)
ypred = classifier.predict(test_data)
accuracy = accuracy_score(test_label, ypred)
error_ratio_bayesian = 1 - accuracy

print("Akurasi ratio untuk metode Bayesian Decision Tree : {:.10f}".format(accuracy))
print("Error ratio untuk metode Bayesian Decision Tree : {:.10f}".format(error_ratio_bayesian))
```
